```python
def verificar_pares(lista1,lista2,inicio,fim):
    
  for acao1,acao2 in zip(lista1,lista2):
    try:
      data1 = yf.download(acao1, inicio, fim, progress=False)
      data2 = yf.download(acao2, inicio, fim, progress=False)
    
      result = pd.merge(data1['Adj Close'], data2['Adj Close'], on='Date')

      x1=result['Adj Close_x']
      y1=result['Adj Close_y']

      coin_result = ts.coint(x1, y1) 

      if coin_result[0] < -3.36685108:
        linreg = sp.linregress(data1.values[:,0], data2.values[:,0],)
        slope = linreg[0]
        intercept = linreg[1]
        residual = y1 - (slope*x1 + intercept)

        zscore = sp.zscore(residual)
        zscore = pd.DataFrame(zscore, index=data1.index)

        deltaresiduo = zscore - zscore.shift(1)
        deltaresiduo = deltaresiduo.dropna()
        deltamedia = deltaresiduo.rolling(20).mean()
        stddelta = deltaresiduo.rolling(20).std()
        deltamedia_stdmais2 = deltamedia + (2*stddelta)
        deltamedia_stdmenos2 = deltamedia - (2*stddelta)
        grafico = pd.concat([deltamedia,deltamedia_stdmais2,deltamedia_stdmenos2,zscore],axis=1, keys=['Deltamedia', ' std + 2', 'std -2', 'Z-Score'])
        if(zscore.iloc[-1,0]>deltamedia_stdmais2.iloc[-1,0]):
          print(f'COMPRA:{acao1} / VENDE:{acao2}')    
        elif(zscore.iloc[-1,0]<deltamedia_stdmenos2.iloc[-1,0]):
          print(f'COMPRA:{acao2} / VENDE:{acao1}')
        else:
          print(f'Não há entrada para {acao1} e {acao2}')
    except:
      print(f"não foi possivel baixar dados de {acao1} ou {acao2}")

```
